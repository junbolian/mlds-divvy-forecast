# Topics in Data Engineering
# Final Project
## Northwestern University Fall 2025

Your final project in data engineering will cover many of the topics in class while giving you the flexibility to get creative. You are to design an end-to-end pipeline that can take in data, store it, and have scripts that investigates the data. Essentially, your project should have the following:

* Read data from an API
* Store API data in a database
* Have a DAG that measures your pipeline from start to finish
* Build a repository that allows new users to run easily
* Use your exploratory data analysis skills to find key insights on your dataset

This is a team task with 4-5 people per group. If you cannot find a group, please let Evan know ASAP.

**Please write to Evan by the end of class on Week 6:**
* The members of your group
* Which API you will be working on

You are welcome to use any API that you'd like, but please approve it with Evan by the end of class during Week 6. There are many public APIs that you can find [here.](https://github.com/public-apis/public-apis) 

**Your final submission will be:**
* A GitHub repo with your code and instructions on how to run
* A PowerPoint presentation detailing your API, your project and your findings
* Presenting your repo and PowerPoint to the class during Week 10

You should be able to run this project now and in 5 years. Have fun with it and explore!
